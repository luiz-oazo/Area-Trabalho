```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error, accuracy_score

X = df.drop(columns=['CLASS_ID', "CLASS"])
y = df['CLASS_ID']

# Divida os dados em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crie o modelo de Random Forest
model = RandomForestClassifier(n_estimators=50, random_state=42)  # Para classificação

# Treine o modelo
model.fit(X_train, y_train)

# Faça previsões no conjunto de teste
y_pred = model.predict(X_test)

# Avalie o modelo (para regressão)
rmse = mean_squared_error(y_test, y_pred, squared=False)
print(f"RMSE: {rmse}")

# Avalie o modelo (para classificação, descomente as linhas abaixo e comente as linhas de regressão acima)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Exiba as primeiras previsões e os valores reais
print("\nPrevisões:")
print(y_pred[:10])
print("\nValores Reais:")
print(y_test.head(10))
```
