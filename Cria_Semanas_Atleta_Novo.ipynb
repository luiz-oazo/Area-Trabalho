```python
#
# Rotina de coloca na semana treinos novos
#
# tem que pluga o banco de dados
#
#
#
#
#

import math

def calculate_week_year_by_date(date_str):
    date = datetime.strptime(date_str, '%Y-%m-%dT%H:%M:%SZ')
    week_year = date.isocalendar()[1]

    return week_year

def acerta_desvio(n, media, desvio, novo_valor):
    nova_media = (n * media + novo_valor) / (n + 1)
    
    ssq = n * desvio**2
    nova_ssq = ssq + (novo_valor - media) * (novo_valor - nova_media)

    novo_desvio = math.sqrt(nova_ssq / (n + 1))
    
    return nova_media, novo_desvio
    
def carrega_treino_novo(atl, semana, novo):
    
    #=================
    
    # Chama dados
    
    if novo:
        # Carrega dataframe com os PR por esporte
        df_week = gera_novo(atl)
    
    # Carregas dados de treino do atleta
    
    atividades = pd.read_csv("strava_chris_1_de_janeiro_2024.csv")
    atividades = atividades[atividades['wasCalculated'].isin([True])].sort_values(["sport_type", "start_date"])
    
    #=================
    
    controle = 0
    atv_anterior = ""
    now = datetime.now()
    semana_atual = now.isocalendar()[1]
    ano_atual = now.year
    semana_util = 0
    semana_util_ant = 0
    
    #=================
    
    for atv in atividades.itertuples():
        # pace_orig = (atv.moving_time * 1000.0) / atv.distance
        if atv_anterior != atv.sport_type:
            if controle > 0:
                semana_util = semana_util + 1
                
                while semana_util <= semana_atual:
                    semana_treino_ant = 'pr'+ str(int(semana_util - 1))
                    qtd_treino_ant = 'tr'+ str(int(semana_util - 1))
                    semana_treino = 'pr'+ str(int(semana_util))
                    qtd_treino = 'tr'+ str(int(semana_util))

                    val_temp = df_week.loc[atl_index, semana_treino_ant] * 1.05
                    qtd_temp = df_week.loc[atl_index, qtd_treino_ant] + 1

                    if val_temp > 100:
                        val_temp = 100

                    if qtd_temp > 7:
                        qtd_temp = 7

                    df_week.at[atl_index, semana_treino] = val_temp
                    df_week.at[atl_index, qtd_treino] = qtd_temp

                    semana_util += 1

            controle = 0
        
        if controle == 0:
            
            if atv.sport_type == 'MountainBikeRide' and int(atv.distance) < 5000: continue
            if atv.sport_type == 'Run' and int(atv.distance) < 2000: continue
            if atv.sport_type == 'Ride' and int(atv.distance) < 5000: continue
            if atv.sport_type == 'Swim' and int(atv.distance) < 800: continue
            if atv.sport_type == 'TrailRun' and int(atv.distance) < 2000: continue
            
            sportx = atv.sport_type
            if atv.sport_type == 'MountainBikeRide': sportx = 'MTB'
            if atv.sport_type == 'Run': sportx = 'RUN'
            if atv.sport_type == 'Ride': sportx = 'ROAD'
            if atv.sport_type == 'Swim': sportx = 'SWIM'
            if atv.sport_type == 'TrailRun': sportx = 'TRAIL'
    
            atl_index = df_week[(df_week['sport'] == sportx) & (df_week['ano'] == ano_atual)].index[0]
            atv_anterior = atv.sport_type
        
        if controle == 0:
            soma_pr = int(0)
            for i in range(52):
                campo = 'pr' + str(int(i+1))
                soma_pr += df_week.loc[atl_index, campo]
            
            controle = 1
            if (soma_pr == 0.0).item():
                semana_util = int(atv.week)
                semana_util_ant = semana_util
                semana_treino = 'pr'+ str(semana_util)
                qtd_treino = 'tr'+ str(semana_util)
                df_week.at[atl_index, semana_treino] = atv.score
                df_week.at[atl_index, qtd_treino] = 3
                df_week.at[atl_index, 'data'] = atv.start_date
                
                nova_media, novo_desvio = acerta_desvio(df_week.loc[atl_index, 'qtd_tr'], df_week.loc[atl_index, 'avg_tr'], df_week.loc[atl_index, 'dsv_tr'], atv.score)
                
                df_week.at[atl_index, 'qtd_tr'] = df_week.loc[atl_index, 'qtd_tr'] + 1
                df_week.at[atl_index, 'avg_tr'] = nova_media
                df_week.at[atl_index, 'dsv_tr'] = novo_desvio
                #print(df_week.loc[atl_index, 'qtd_tr'], df_week.loc[atl_index, 'avg_tr'], df_week.loc[atl_index, 'dsv_tr'])
                
                continue

        semana_util = int(atv.week)
        semana_treino = 'pr'+ str(semana_util)
        qtd_treino = 'tr'+ str(semana_util)
        
        if semana_util > semana_util_ant:
            semana_util_ant = semana_util_ant + 1
                
            while semana_util_ant <= semana_util:
                semana_treino_ant = 'pr'+ str(int(semana_util_ant - 1))
                qtd_treino_ant = 'tr'+ str(int(semana_util_ant - 1))
                semana_treino = 'pr'+ str(int(semana_util_ant))
                qtd_treino = 'tr'+ str(int(semana_util_ant))

                val_temp = df_week.loc[atl_index, semana_treino_ant] * 1.05
                qtd_temp = df_week.loc[atl_index, qtd_treino_ant] + 1

                if val_temp > 100: val_temp = 100
                if qtd_temp > 7: qtd_temp = 7

                df_week.at[atl_index, semana_treino] = val_temp
                df_week.at[atl_index, qtd_treino] = qtd_temp

                semana_util_ant += 1
        
        semana_util_ant = semana_util
        
        min_pace = 0.0
        max_pace = 0.0
        if df_week.loc[atl_index, 'qtd_tr'] >= 3:
            min_pace = df_week.loc[atl_index, 'avg_tr'] - (2 * df_week.loc[atl_index, 'dsv_tr'])
            max_pace = df_week.loc[atl_index, 'avg_tr'] + (2 * df_week.loc[atl_index, 'dsv_tr'])
            
            # print(min_pace, atv.score, max_pace)
            if min_pace > atv.score or atv.score > max_pace: continue
        
        val_calc = df_week.loc[atl_index, semana_treino]
        qtd_calc = df_week.loc[atl_index, qtd_treino] - 1
        
        if qtd_calc < 3: qtd_calc = 3
        
        #print(semana_util, 'antes', atv.score, val_calc, qtd_calc, min_pace, pace_orig, max_pace)
        
        val_calc = (atv.score + val_calc * qtd_calc) / (qtd_calc + 1)

        #print(semana_util, 'depois', atv.score, val_calc, qtd_calc)        
        
        if val_calc < 1: val_calc = 1
        if val_calc > 100: val_calc = 100

        df_week.at[atl_index, semana_treino] = val_calc
        df_week.at[atl_index, qtd_treino] = qtd_calc
        
        df_week.at[atl_index, 'data'] = atv.start_date
                
        nova_media, novo_desvio = acerta_desvio(df_week.loc[atl_index, 'qtd_tr'], df_week.loc[atl_index, 'avg_tr'], df_week.loc[atl_index, 'dsv_tr'], atv.score)
                
        df_week.at[atl_index, 'qtd_tr'] = df_week.loc[atl_index, 'qtd_tr'] + 1
        df_week.at[atl_index, 'avg_tr'] = nova_media
        df_week.at[atl_index, 'dsv_tr'] = novo_desvio
        #print(min_pace, atv.score, max_pace)
        
    if controle > 0:
        semana_util = semana_util + 1
                
        while semana_util <= semana_atual:
            semana_treino_ant = 'pr'+ str(int(semana_util - 1))
            qtd_treino_ant = 'tr'+ str(int(semana_util - 1))
            semana_treino = 'pr'+ str(int(semana_util))
            qtd_treino = 'tr'+ str(int(semana_util))

            val_temp = df_week.loc[atl_index, semana_treino_ant] * 1.05
            qtd_temp = df_week.loc[atl_index, qtd_treino_ant] + 1

            if val_temp > 100:
                val_temp = 100
            
            if qtd_temp > 7:
                qtd_temp = 7
                
            df_week.at[atl_index, semana_treino] = val_temp
            df_week.at[atl_index, qtd_treino] = qtd_temp

            semana_util += 1

    return df_week
```
