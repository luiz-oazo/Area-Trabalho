```python
import oracledb
import json
import pandas as pd
```

```python
import json

def conexao():
    DATA_FILE = "oracle_conn.json"
    file = open(DATA_FILE, 'r')
    xfile = json.load(file)

    connection = oracledb.connect(
        user=xfile[0]["user"],
        password=xfile[0]["password"],
        dsn=xfile[0]["dsn"] 
    )
    return(connection)
```

```python
connection = conexao()
cursor = connection.cursor()

# Executando a consulta para listar as tabelas do esquema atual
cursor.execute("SELECT table_name FROM user_tables")

# Buscando e imprimindo todas as tabelas
tables = cursor.fetchall()
tables
```

```python
connection = conexao()
cursor = connection.cursor()

create_table_query = '''
CREATE TABLE filmes (
    id NUMBER GENERATED BY DEFAULT AS IDENTITY PRIMARY KEY,
    filme VARCHAR2(50),
    ano NUMBER,
    genero VARCHAR2(50),
    ator VARCHAR2(50),
    atriz VARCHAR2(50)
)
'''

cursor.execute(create_table_query)
print("Tabela funcionários criada com sucesso.")
```

```python
import requests

# GET request
response = requests.get('http://127.0.0.1:5000/filmes')
print(response.json())
```

```python
# GET request
response = requests.get('http://127.0.0.1:5000/filmes/42')
print(response.json())
```

```python
busca = {'ator': 'Leonardo DiCaprio'}
response = requests.get('http://127.0.0.1:5000/filmes', params=busca)
for i in response.json():
    print(i)
```

```python
# GET request produto
response = requests.delete('http://127.0.0.1:5000/filmes/41')
print(response.json())
```

```python
novo_filme = {"filme": "Titanic","ano": 1997,"genero": "Romance/Drama","ator": "Leonardo DiCaprio","atriz": "Kate Winslet"}
response = requests.post('http://127.0.0.1:5000/filmes', json=novo_filme)
print(response.json())
```

```python
updade_produto = {"ano": 2023, "ator": "Natan"}
response = requests.put('http://127.0.0.1:5000/filmes/42', json=updade_produto)
print(response.json())
```
