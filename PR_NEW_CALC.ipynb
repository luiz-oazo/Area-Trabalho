```python
def pr_treino(**treino):
    
    bonus = 0.0
    
    for chave, val_atl in treino.items():
        if chave != "tipo" and chave != "pace":
            
            if pr_bonus.get(treino['tipo']):
            
                if ((pr_bonus[treino['tipo']][chave]['tipo'] == 'fixo') and 
                    (pr_bonus[treino['tipo']][chave]['valor'] == val_atl)):
                    bonus += pr_bonus[treino['tipo']][chave]['bonus']
                    print(chave, pr_bonus[treino['tipo']][chave], bonus)
                
                if pr_bonus[treino['tipo']][chave]['tipo'] == 'soma':
                    bon_tmp = pr_bonus[treino['tipo']][chave]['bonus']
                    inicio = pr_bonus[treino['tipo']][chave]['inicio']
                    valor = pr_bonus[treino['tipo']][chave]['valor']
                    
                    if valor:
                        bonus += bon_tmp * ((val_atl - inicio) / valor)
                    print(chave, pr_bonus[treino['tipo']][chave], (bon_tmp * ((val_atl - inicio) / valor)), bonus)
                
    if pr_calc.get(treino['tipo']):
        pace_calca = pr_calc[treino['tipo']].get('pace')
        pts_calca = pr_calc[treino['tipo']].get('pts')
        
        print(treino['pace'], pace_calca, pts_calca, bonus)
        
        calcula = (treino['pace'] - pace_calca) * (1 - bonus) * pts_calca
        
        if calcula < 1: calcula = 1
        if calcula > 100: calcula = 100
            
    return calcula
```

```python
import math
def acerta_desvio(n, media, desvio, novo_valor):
    # Calcular a nova média
    nova_media = (n * media + novo_valor) / (n + 1)
    
    # Calcular a soma dos quadrados das diferenças para o desvio padrão atual
    ssq = n * desvio**2
    
    # Calcular a nova soma dos quadrados das diferenças
    nova_ssq = ssq + (novo_valor - media) * (novo_valor - nova_media)
    
    # Calcular o novo desvio padrão
    novo_desvio = math.sqrt(nova_ssq / (n + 1))
    
    return nova_media, novo_desvio
```
